# Step 4: Pre-Process Dataset

```python
# Pre-Processing

data.tail()
data.drop(data.loc[:, 'iresp.1':'iresp.170'], inplace = True, axis = 1) 
data_final = data
```

```python
# Count Null Values

print(data_final.isnull().sum())
print(data_final.shape)
```

```python
# Drop Null Values

data_final.dropna(inplace=True)
print(data_final.isnull().sum())
print(data_final.shape)
```

```python
# Time

data_final['mean_idur'] = data_final.loc[:, 'idur.1':'idur.170'].mean(axis=1)
data_final['median_idur'] = data_final.loc[:, 'idur.1':'idur.170'].median(axis=1)
data_final['max_idur'] = data_final.loc[:, 'idur.1':'idur.170'].max(axis=1)
data_final['min_idur'] = data_final.loc[:, 'idur.1':'idur.170'].min(axis=1)

# Correct Answers

data_final['iraw_score'] = data_final.loc[:, 'iraw.1':'iraw.170'].sum(axis=1)
data_final['iraw_score_pct'] = data_final['iraw_score']/170
data_final
```

# Step 5: Feature Selection

```python
# iraw_idur_OEF (both+OEF)

col_features = ['Attempt', 'tot_time', 'mean_idur', 'median_idur', 'max_idur', 'min_idur', 'iraw_score_pct']

for x in range(1,171):
    col_features.append('idur.'+str(x))

for i in range(1,171):
    col_features.append('iraw.'+str(i))

X = data_final.loc[:, col_features]
Y = data_final.Flagged    
```

## Step 8.1: Chi2 Features Selection

<div class="alert alert-warning">
<b>This score can be used to select the n_features features with the highest values for the test chi-squared statistic from X, which must contain only non-negative features such as booleans or frequencies (e.g., term counts in document classification), relative to the classes.</b>
</div>

```python
# Import Libraries
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

# Apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,Y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)

# naming the dataframe columns
featureScores.columns = ['Feature','Score']  
```

```python
featureScores
```

```python
# Print 10 Best Features

best = pd.DataFrame(featureScores.nlargest(10,'Score'))
best["Ranking"] = ["       --->      Rank-01","       --->      Rank-02","       --->      Rank-03","       --->      Rank-04","       --->      Rank-05","       --->      Rank-06","       --->      Rank-07","       --->      Rank-08","       --->      Rank-09","       --->      Rank-10"]
print("\n==================================================\n")
print("================= TOP 10 FEATURES ================")
print("\n==================================================\n")
print(best.to_string(index=False))
print("\n==================================================\n")
```

### Features Set-1

```python
Features_Set1 = X[best["Feature"]]

print("\n\nAutomatic Selected Features Set 1:")
print("==================================\n")
Features_Set1
```

## Step 8.2: Extra Tree Features Selection

<div class="alert alert-warning">
<b>This class implements a meta estimator that fits a number of randomized decision trees (a.k.a. extra-trees) on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.</b>
</div>

```python
# Import Libraries
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier()
model.fit(X,Y)
columns = X.columns.values

print(model.feature_importances_)
feat_importances = pd.Series(model.feature_importances_)
```

```python
# Print 10 Best Features

top = feat_importances.nlargest(10)
feature = top.index
names = columns[feature]

top_features = pd.DataFrame()
top_features["Features"]=names
top_features["    Score"]=top.values
top_features["Ranking"] = ["       --->      Rank-01","       --->      Rank-02","       --->      Rank-03","       --->      Rank-04","       --->      Rank-05","       --->      Rank-06","       --->      Rank-07","       --->      Rank-08","       --->      Rank-09","       --->      Rank-10"]
print("\n===================================================\n")
print("================== TOP 10 FEATURES ================")
print("\n===================================================\n")
print(top_features.to_string(index=False))
print("\n===================================================\n")
```

### Features Set-2

```python
Features_Set2 = X[top_features["Features"]]

print("\n\nAutomatic Selected Features Set 2:")
print("==================================\n")
Features_Set2
```

## Step 8.3: REF Features Selection

<div class="alert alert-warning">
<b>RFE is popular because it is easy to configure and use and because it is effective at selecting those features (columns) in a training dataset that are more or most relevant in predicting the target variable.</b>
</div>

```python
# Import Libraries
from sklearn.feature_selection import RFE

rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=1)
fit= rfe.fit(X, Y)
columns = X.columns.values
top = []
ranking = []

for i in range(X.shape[1]):
    if(rfe.ranking_[i]<=10):
        top.append(columns[i])
        ranking.append(rfe.ranking_[i])
```

```python
# Print 10 Best Features

top_features = pd.DataFrame()
top_features["Features"]= top
top_features["                   Ranking"] = ranking
top_features = top_features.sort_values('                   Ranking')
print("\n==============================================\n")
print("=============== TOP 10 FEATURES ==============")
print("\n==============================================\n")
print(top_features.to_string(index=False))
print("\n==============================================\n")
```

### Features Set-3

```python
Features_Set3 = X[top_features["Features"]]

print("\n\nAutomatic Selected Features Set 3:")
print("==================================\n")
Features_Set3
```

## Step 8.4: Pearson Correlation Features Selection

<div class="alert alert-warning">
<b>Pearson's Correlation method is used for finding the association between the continuous features and the class feature.</b>
</div>

```python
# Find Correlation

C = X.copy()
C["Output"] = Y
cor = C.corr()

print("\n\nPearson Correlation:")
print("====================\n")
cor
```

```python
# Correlation with output variable
cor_target = abs(cor["Output"])

# Selecting highly correlated features
relevant_features = cor_target[cor_target>0.1]
relevant_features
```

### Features Set-4

```python
a = np.array(relevant_features.index)
a = list(a)
a.remove("Output")
a.remove("min_idur")

Features_Set4 = X[a]

print("\n\nAutomatic Selected Features Set 4:")
print("==================================\n")
Features_Set4
```
